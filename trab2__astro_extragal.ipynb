```
#antes de rodar, este arquivo deve ser colocado nesse programa
df = pd.read_table('Fis2013a.dat', delim_whitespace=True, skiprows =1, header = None)
df = df.rename(columns={0:'R',1:'B',2:'sigB',3:'V',4:'sigV',5:'R',6:'sigR',7:'(B-V)',8:'s(B-V)',9:'(B-R)',10:'s(B-R)',11:'(V-R)',12:'s(V-R)'})

df2 = pd.read_table('bvr_008_5000Myr.dat', delim_whitespace=True, header = None, skiprows =1)
df2 = df2.rename(columns={0:'M(mo)',1:'B',2:'V',3:'R',4:'(B-V)',5:'(B-R)',6:'(V-R)'})

df2['(B-V)'] = df2['(B-V)']+0.25
df2['B'] = df2['B']+10.25

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(df['(B-V)'],df['B'], ".", label = 'Aglomerado Simulado Fis2013a')
ax.plot(df2['(B-V)'],df2['B'], ".", label = 'Isócrona')
ax.set_title('Diagrama Cor Magnitude (CMD)')
fig.gca().invert_yaxis()
ax.legend()
ax.set_xlabel("(B-V)")
ax.set_ylabel(r"$M_B$")
fig.savefig("CMDa")
fig.show() 
```
